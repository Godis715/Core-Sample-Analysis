```python
from oil_carbon_detection.areas_detection import get_areas
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
dl_img = plt.imread('data/Unload1/data/1000196.jpeg')
uv_img = plt.imread('data/Unload1/data/1000197.jpeg')
```

```python
(oil_area, oil_dict), (carbon_area, carbon_dict) = get_areas(dl_img, uv_img)
```

```python
masks = np.array([[[carbon_area[i][j]*(-100), oil_area[i][j]*(-100), 0] 
                            for j in range(len(uv_img[0]))] for i in range(len(uv_img))])
```

```python
fig, (ax1,ax2,ax3) = plt.subplots(1, 3,  figsize=(10,100), sharex=True, sharey=True)

ax1.imshow(dl_img)
ax1.set_title('DL')
ax2.imshow(uv_img)
ax2.set_title('UV')
ax3.imshow(masks)
ax3.set_title('Final mask')

plt.show()
```
